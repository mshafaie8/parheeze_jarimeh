```python
import pandas as pd
import tabula as tb
import numpy as np
```

```python
data_ag = tb.read_pdf("data/sweepingschedag.pdf", pages = [1, 2, 3, 4, 5, 6, 7])
data_hz = tb.read_pdf("data/sweepingschedhz.pdf", pages = [1, 2, 3, 4, 5, 6, 7, 8])
data_nums = tb.read_pdf("data/sweepingschednums.pdf", pages = 1)
```

```python
data_ag[0]
```

```python
cleaned_data = []
```

## Deal with A-G Data

### First Page

```python
data_ag[0].head()
```

```python
fixed_ag0 = data_ag[0].iloc[1:, :]
```

```python
fixed_ag0.head()
```

```python
which_day = fixed_ag0["Unnamed: 0"].str.split().str[1] + fixed_ag0["Unnamed: 0"].str.split().str[0]
```

```python
fixed_ag0["Day of Month"] = which_day + " " + fixed_ag0["Day of"]
```

```python
fixed_ag0 = fixed_ag0[["Street Name", "From", "To", "Day of Month", "AM"]]
```

```python
fixed_ag0.rename({"AM": "Time"}, axis = 1, inplace = True)
```

```python
fixed_ag0.head()
```

```python
cleaned_data.append(fixed_ag0)
```

### Deal With Pages 2 - 7

```python
def process_data_ag(old_data):
    old_data.rename({"AM\rPM": "Time"}, axis = 1, inplace = True)
    split = old_data['Day of\rMonth'].str.split("\r")
    day = split.str[1].str[1:]
    suffix = split.str[0]
    num = split.str[1].str[0]
    old_data["Day of Month"] = num + suffix + " " + day
    return old_data[["Street Name", "From", "To", "Day of Month", "Time"]]
```

```python
for page_num in np.arange(1, 7):
    processed = process_data_ag(data_ag[page_num])
    cleaned_data.append(processed)
```

## Deal with H-Z Data

```python
for page_num in range(len(data_hz)):
    og = data_hz[page_num]
    if page_num == 0:
        new = og.iloc[1:, :][["Street Name", "From", "To", "Day of", "AM"]]
        new.rename({"Day of": "Day of Month", "AM": "Time"}, axis = 1, inplace = True)
        cleaned_data.append(new)
    else:
        new = og[["Street Name", "From", "To", "Day of\rMonth", "AM\rPM"]]
        new.rename({"Day of\rMonth": "Day of Month", "AM\rPM": "Time"}, axis = 1, inplace = True)
        cleaned_data.append(new)
```

## Deal with Nums Data

```python
cleaned_nums = data_nums[0].iloc[1:, :].rename({"Day of": "Day of Month", "AM": "Time"}, axis = 1)
cleaned_nums = cleaned_nums[["Street Name", "From", "To", "Day of Month", "Time"]]
cleaned_nums.head()
```

```python
cleaned_data.append(cleaned_nums)
```

## Put all the DataFrames together

```python
full_cleaned_data = pd.concat(cleaned_data, ignore_index = True)
```

```python
def just_street_name_lowercase(street):
    street_type = ['ct', 'circle', 'crescent', 'pl', 'ave', 'dr', 'terr', 
                    'blvd', 'cresc', 'place', 'real', 'rd', "street"]
    street_name = street.lower()
    split = street_name.split()
    if split[-1] in street_type:
        result = split[0]
        for phrase in split[1:-1]:
            result += (" " + phrase)
        return result
    return street_name
```

```python
full_cleaned_data["Street Name"] = full_cleaned_data["Street Name"].apply(just_street_name_lowercase)
```
